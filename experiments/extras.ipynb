```python
import cv2
```

```python
from matplotlib.pyplot import imshow
from matplotlib import pyplot as plt
```

# Background

```python
bg1 = cv2.imread('../assets/background.png')
bg2 = cv2.imread('../assets/background3.png')
bg3 = cv2.imread('../assets/background2.png')
```

```python
plt.rcParams['figure.figsize'] = [15, 13]

f, axarr = plt.subplots(1,3)

# plt.figure()

axarr[0].imshow(bg1)
axarr[0].set_title('Construction')
axarr[1].imshow(bg2)
axarr[1].set_title('Empty Stadium')
axarr[2].imshow(bg3)
axarr[2].set_title('Crowd')
```
