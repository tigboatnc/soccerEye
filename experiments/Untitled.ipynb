```python
cap = cv2.VideoCapture(str(dataset_files[0]))
```

```python
total_frames = cap.get(7)
print(f'Total Frames in Video: {total_frames}')
```

```python
# Display Random Frame 
frame_no = 50
if frame_no < int(total_frames):
    cap.set(1, frame_no)
    ret, frame = cap.read()
    imshow(frame)
else:
    print('Frame Number selected larger than max frame')
```

```python
# HSV Frame 
frame_hsv = cv2.cvtColor(frame, cv2.COLOR_BGR2HSV)
# imshow(frame_hsv)

fig, axs = plt.subplots(3)
fig.suptitle('HSV')
axs[0].imshow(frame_hsv[:,:,0])
axs[1].imshow(frame_hsv[:,:,1])
axs[2].imshow(frame_hsv[:,:,2])
```
